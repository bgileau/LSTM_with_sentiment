```python
# import items for vader NLK NLP algorithm
import pandas as pd
import numpy as np
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
```

```python
df = pd.read_csv('Combined_News_DJIA.csv')

# Download the vader_lexicon
nltk.download('vader_lexicon')

# Instantiate the sentiment intensity analyzer with the existing lexicon
vader = SentimentIntensityAnalyzer()
```

```python
# Define the headlines' columns names
headlines = ['Top'+str(i) for i in range(1,26)]

# Fill NaN values with empty strings
df[headlines] = df[headlines].fillna('')

# Calculate sentiment scores for each headline and then average them for each day
df['average_sentiment'] = df[headlines].applymap(vader.polarity_scores).applymap(lambda x: x['compound']).mean(axis=1)
```

```python
# compute vader score for string x
# x = "Russia 'ends Georgia operation'"
x = 'China bans youtube'
print(vader.polarity_scores(x))
```
